```python



def dataframe_detailed(df):


    """
    print(f"DATAFRAME SHAPE: {df.shape}\n\n")
    print(f"{df.info()}\n\n")

    print(f"DATAFRAME HEAD:\n {df.head()}\n\n")

    print(f"DATAFRAME KEY STATISTIC DESCRIPTION:\n {df.describe()}\n\n")



def check_outliers(data, columns):

 for column in columns:

    # Calculate IQR (Interquartile Range)
    iqr = data[column].quantile(0.75) - data[column].quantile(0.25)

    # Define lower and upper thresholds
    lower_threshold = data[column].quantile(0.25) - 1.5 * iqr
    upper_threshold = data[column].quantile(0.75) + 1.5 * iqr

    # Find outliers
    outliers = data[(data[column] < lower_threshold) |

                    (data[column] > upper_threshold)]

    # Print the count of outliers

    print(f"{column}\nNumber of outliers: {len(outliers)}\n")
    

def replace_value(df, col, old_value, new_value):


    if df[col].eq(old_value).any():

        df[col] = df[col].replace({old_value: new_value})
        


def create_pie_chart(data, column, ax, title):

    values = data[column].value_counts()

    ax.pie(values, labels=values.index, autopct='%1.1f%%', startangle=90)

    ax.set_title(title)
    

def checkVIF(X):

    vif = pd.DataFrame()

    vif['Features'] = X.columns

    vif['VIF'] = [variance_inflation_factor(

        X.values, i) for i in range(X.shape[1])]

    vif['VIF'] = round(vif['VIF'], 2)

    vif = vif.sort_values(by="VIF", ascending=False,)
    return (vif)



def calculate_vif(data_frame):

    vif_data = pd.DataFrame()

    vif_data["Variable"] = data_frame.columns

    vif_data["VIF"] = [variance_inflation_factor(

        data_frame.values, i) for i in range(data_frame.shape[1])]
    return vif_data



def perform_grid_search(model, param_grid, X, y, cv):

    grid_search = GridSearchCV(

        estimator=model, param_grid=param_grid, cv=cv, scoring='accuracy', n_jobs=-1
    )

    grid_search.fit(X, y)


    best_params = grid_search.best_params_

    best_score = grid_search.best_score_


    return best_params, best_score


def plot_roc_curve(model, X_test, y_test):
    """
    Plot the ROC curve for a given model.

    Parameters:
    - model: The trained machine learning model.
    - X_test: The feature matrix for the test set.
    - y_test: The true labels for the test set.
    """
    # Predict probabilities for positive class
    y_prob = model.predict_proba(X_test)[:, 1]

    # Calculate ROC curve
    fpr, tpr, thresholds = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    # Plot ROC curve
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2,
             label='ROC curve (area = {:.2f})'.format(roc_auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.show()
    

def oversample_with_smote(X_train, y_train):
    print("Before OverSampling - counts of label '1': {}".format(sum(y_train == 1)))
    print("Before OverSampling - counts of label '0': {} \n".format(sum(y_train == 0)))
    print("*******************************************************************")

    sm = SMOTE(random_state=100)
    X_train_smote, y_train_smote = sm.fit_resample(X_train, y_train.ravel())

    print('After OverSampling with SMOTE - x_train: {}'.format(X_train_smote.shape))
    print('After OverSampling with SMOTE - y_train: {} \n'.format(y_train_smote.shape))

    print("After OverSampling with SMOTE - counts of label '1': {}".format(sum(y_train_smote == 1)))
    print("After OverSampling with SMOTE - counts of label '0': {}".format(sum(y_train_smote == 0)))

    return X_train_smote, y_train_smote
```

## Data Preprocessing

Data preprocessing is a vital step in the data analysis and machine learning process. Its main objectives include handling missing data through imputation or removal, managing outliers to prevent skewed results, encoding categorical variables into a numerical format suitable for analysis, and scaling or normalizing features to ensure a consistent scale. Preprocessing also addresses imbalanced data distribution, incorporates feature engineering for model improvement, ensures data consistency, and reduces dimensionality for computational efficiency. Ultimately, the goal of data preprocessing is to prepare raw data for comprehensive analysis, exploration, and effective training of machine learning models, leading to improved model performance and accurate predictions.

```python
df_dummy = pd.get_dummies(df, prefix=['Contract', 'PaymentMethod'], columns=['Contract', 'PaymentMethod'], drop_first=True)

df_dummy.head()
```


In the preceding code, we've undertaken essential preprocessing steps to transform categorical variables into a format that is compatible with machine learning models.

Goal: The objective is to handle columns with multiple categories by creating binary columns for each unique category, a process known as One-Hot Encoding.
Process: For the specified columns ('Contract' and 'PaymentMethod'), new binary columns are generated for each category, indicating the presence or absence of that category using 1s and 0s.
Reasoning: One-Hot Encoding preserves detailed information about each category, allowing the machine learning model to learn distinct patterns associated with different categories.

Consideration: However, it's important to drop the first level of each categorical feature to mitigate multicollinearity, a situation where features are highly correlated, which can affect model performance

```python
categorical_feature = df.dtypes == object
final_categorical_feature = df.columns[categorical_feature].tolist()
final_categorical_feature = [
    item for item in final_categorical_feature if item not in ('Contract', 'PaymentMethod')]

```

```python
le = LabelEncoder()
df_dummy[final_categorical_feature] = df_dummy[final_categorical_feature].apply(
    lambda col: le.fit_transform(col))
df_dummy.head(5)
```

Before doing any modelling we check for multicollinearity among the features in the data set.

The correlation information above is vital since it will be combined with the multicollinearity which will enable us to be able to remove it

```python
calculate_vif(df_dummy)
```

The columns Tenure,TotalCharges,PhoneService and MonthlyCharges have very high degree of multicollinearity. 

This means that the columns have  strong  relationship with other independent variables in the model, and its variance in the classification coefficients is inflated due to this correlation. 

High multicollinearity can lead to inflated standard errors and unstable coefficient estimates in logistic regression. This, in turn, may affect the interpretability of the model.

To be able to remove multicollinearity we will combine the highly correlated features which are "tenure" and "TotalCharges" into a new feature representing the average monthly charges. This could capture the overall financial contribution without having two highly correlated features.

```python
# Create a new feature
df_dummy['AvgMonthlyCharges'] = df_dummy['TotalCharges'] / df_dummy['tenure']

# Drop the original features
df_dummy = df_dummy.drop(['TotalCharges', 'tenure'], axis=1)
```

In a machine learning classification problem, class imbalance can pose challenges for model training and evaluation. When one class is significantly underrepresented, the model may become biased towards the majority class, leading to poor performance on the minority class.

To address this issue, one common technique is SMOTE (Synthetic Minority Over-sampling Technique), where synthetic instances are generated for the minority class to create a more balanced distribution.

The reason for choosing SMOTE over other methods is that it is applicable to small datasets, providing a more diverse representation of the minority class. Additionally, SMOTE introduces synthetic instances, addressing potential issues related to duplicating existing data points. While it involves a slightly more complex mechanism than basic oversampling, its effectiveness in handling imbalanced datasets, especially with limited data, makes it a valuable choice.

```python
# Initialize the scaler
scale = StandardScaler()

# Get the column names of the features
col = X_train.columns

# Apply standard scaling to the features in both training and testing sets
X_train[col] = scale.fit_transform(X_train[col])
X_test[col] = scale.transform(X_test[col])
```

```python
X_train_smote = X_train.copy()
y_train_smote = y_train.copy()

X_train_smote, y_train_smote = oversample_with_smote(
    X_train_smote, y_train_smote)
```

SMOTE has been applied to address the class imbalance, and the output demonstrates the balanced distribution achieved after oversampling, paving the way for more robust training of classification models, especially for tasks like predicting customer churn.

Next we fit our baseline Logistic Regression Model

```python

# #baseline model
# Create a pipeline with preprocessing and logistic regression model
lr_pipeline = Pipeline(steps=[
    ('classifier', LogisticRegression(random_state=12,solver='newton-cg'))
])
# Fit the pipeline on the training data
lr_pipeline.fit(X_train_smote, y_train_smote)

y_pred_lr = lr_pipeline.predict(X_test)

# Print the full classification report
print("Classification Report for Logistic Regression Model:\n")
print(classification_report(y_test,y_pred_lr))

#ROC Curve
plot_roc_curve(lr_pipeline, X_test, y_test)
conf_matrix = confusion_matrix(y_test, y_pred_lr)
# Display confusion matrix as a heatmap
disp = ConfusionMatrixDisplay(
    confusion_matrix=conf_matrix, display_labels=lr_pipeline.classes_)
disp.plot(cmap='Blues')
```

## Random Forest Classification Model

```python
model_rf = RandomForestClassifier(random_state=50)

param_grid = {
    'n_estimators': [10, 25, 50, 100],
    'min_samples_split': [2, 4, 6, 8],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}
CV = StratifiedKFold(n_splits=10, random_state=0, shuffle=True)

best_params_rf,best_score_rf = perform_grid_search(model = model_rf,param_grid=param_grid,X=X_train_smote,y= y_train_smote,cv=CV)

print("Best Parameters:", best_params_rf)
print("Best Accuracy:", best_score_rf)
```

```python
# Create the RandomForestClassifier with the best parameters
RF = RandomForestClassifier(**best_params_rf, random_state=42)
#fit the tuned model
RF.fit(X_train_smote, y_train_smote)

#the cross_vals
cross_val_RF = cross_val_score(
    RF, X_train_smote, y_train_smote, cv=CV, scoring='accuracy')

cross_val_RF_f1 = cross_val_score(RF, X_train_smote, y_train_smote, cv=CV,
                                  scoring='f1')


# Cross-validated accuracy scores
print("Mean accuracy:", cross_val_RF.mean())

# Cross-validated F1 scores
print("Mean F1 score:", cross_val_RF_f1.mean())

# Print the classification report
print("Classification Report For Random Forest:\n", classification_report(
    y_test, y_pred=RF.predict(X_test)))

# plot the ROC Curve
plot_roc_curve(RF, X_test, y_test)

conf_matrix = confusion_matrix(y_test, y_pred=RF.predict(X_test))
# Display confusion matrix as a heatmap
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=RF.classes_)
disp.plot(cmap='Blues')
```

**F1-Score:**

- For Churn (Class 1): 53% F1-score, reflecting the balance between precision and recall for churn instances.

**Weighted Average Consideration:**
